```
"""
truthTableEvaluator
"""

# Given the following inputs:


# P = # True or False
# Q = # True or False
# op = # '^' (logical AND, conjunction)
#      # OR, 'v' (logical OR, disjunction)
#      # OR, '->' (logical conditional, implication)
#      # OR, '<->' (biconditional)
# determine the correct outcome.

# https://medium.com/i-math/intro-to-truth-tables-boolean-algebra-73b331dd9b94

#P / Q // P and Q / 

index = [0,1,2,3]
op = ['^','v','->','<->']
ValuesP = [True, True, False, False]
ValuesQ = [True, False, True, False]
PandQ = []
PorQ = []
PimpliesQ = []
PiffQ = []


#SUBROUTINE 1
for x in index:
  P = ValuesP[x]
  Q = ValuesQ[x]
  output1 = P and Q
  PandQ.append(output1)
  output2 = P or Q
  PorQ.append(output2)
  output3 = (not P) or Q
  PimpliesQ.append(output3)
  output4 = ((not P) or Q) and ((not Q) or P)
  PiffQ.append(output4)

print('P\tQ\tP ^ Q\tP v Q\tP -> Q\tP <-> Q')
for x in index:
  print(str(ValuesP[x])+'\t'+str(ValuesQ[x])+'\t'+str(PandQ[x])+'\t'+str(PorQ[x])+'\t'+str(PimpliesQ[x])+'\t'+str(PiffQ[x]))
print('\n==================================\n\n')


#subroutine2
op = input("""Enter an operator the valid operators are \n '^' and\n'v' or \n'->' conditional \n'<-> biconditional\n'""")
for x in index:
  if op == '^':
    output = P and Q
    print(output)
  elif op == 'v':
    output = P or Q
    PorQ.append(output)
    #print(output)
  elif op == '->':
    output = (not P) or Q
    PimpliesQ.append(output)
    #print(output)
  elif op == '<->':
    output = ((not P) or Q) and ((not Q) or P)
    PiffQ.append(output)
    #print(output)
  else:
    print('Invalid operator')

print('\n\n====================\n\n')
#SUBROUTINE 3
if op == '^':
  print(PandQ)
elif op == 'v':
  print(PorQ)
elif op == '->':
  print(PimpliesQ)
elif op == '<->':
  print(PiffQ)
else:
  print('Invalid operator')
```
