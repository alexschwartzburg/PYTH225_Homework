```
"""
Control Flow II

Write a Python program that prints all the numbers from 
0 to 6 except 3 and 6. Note: Use a 'continue' statement. 
"""

#rules based output
print('rules based output') #this will skip 0.
for x in range(0,6):
  if x % 3 == 0:
    continue
  else:
    print(x)

#exceptions based output
print('\n\nexceptions based output') #this won't skip 0.
for x in range(0,6):
  if x == 3 or x ==6:
    continue
  else:
    print(x)

print('\n\nThe exceptions based approach is correct because it doesn\'t skip 0')
```

```
"""
GCD
"""

# Find the greatest common denominator (GCD) of two number input by a user. 
# Then print out 'The GCD of <first number> and <second number> is <your result>.'

#print('Enter two numbers to find their greatest common denominator.')

user_input1 = input('First number: ')
user_input2 = input('Second number: ')


def intersection(factors_a,factors_b):
  factors = []
  for x in factors_a:
    if x in factors_b:
      factors.append(x)
    else:
      continue
  return factors

def GCD(factors):
  try:
    valid_factors = []
    for x in factors:
      if x == 1:
        pass
      else:
        valid_factors.append(x)
    return max(valid_factors)
  except Exception:
    print('Error: 1 isn\'t allowed to be the GCD two numbers.')

def LCD(factors):
  try:
    valid_factors = []
    for x in factors:
      if x == 1:
        pass
      else:
        valid_factors.append(x)
    return min(valid_factors)
  except Exception:
    print('Error: 1 isn\'t allowed to be the LCD two numbers.')


a = int(user_input1)
b = int(user_input2)

factors_a = []
factors_b = []

for x in range(1,a+1):
  if a % x == 0:
    factors_a.append(x)
print('factors_a = '+str(factors_a))

for x in range(1,b+1):
  if b % x == 0:
    factors_b.append(x)
print('factors_b = '+str(factors_b))

factors = intersection(factors_a,factors_b)
print('common factors '+str(factors))
GCD = GCD(factors)
LCD = LCD(factors)
print('Greatest Common Denominator '+str(GCD)+'\n',
      'Least Common Demoninator '+str(LCD))

if GCD == None:
    print('There is no GCD of 187 and 987.')
else:
    print(f'The GCD of {user_input1} and {user_input2} is {GCD}.')
if LCD == None:
    print('There is no LCD of 187 and 987.')
else:
    print(f'The LCD of {user_input1} and {user_input2} is {LCD}.')


# I changed the problem a little bit because functions made it easier to accomplish.
```
