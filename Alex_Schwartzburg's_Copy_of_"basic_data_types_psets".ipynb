### P1.PY

```
"""
Typcasting w. Integers & Floats
"""

# Convert these numbers into floats and back. Print out each result as well as its data type.

five = 5
zero = 0
neg_8 = -8
neg_22 = -22

a = float(five)
b = float(zero)
c = float(neg_8)
d = float(neg_22)

e = int(a)
f = int(b)
g = int(c)
h = int(d)

print(str(e)+" "+str(type(e)))
print(str(f)+" "+str(type(f)))
print(str(g)+" "+str(type(g)))
print(str(h)+" "+str(type(h)))
```

```
"""
Typcasting w. Strings
"""

# Convert these variables into strings and then back to their original data types. Print out each result as well as its data type. What do you notice about the last one?

five = 5
zero = 0
neg_8 = -8
T = True
F = False

list1 = [five,zero,neg_8,T,F]

for x in list1:
  print(str(x)+str(type(x)))
print('\n')

five = str(five)
zero = str(zero)
neg_8 = str(neg_8)
T = str(T)
F = str(F)

list2 = [five,zero,neg_8,T,F]

for x in list2:
  print(str(x)+str(type(x)))
print('\n')

five = int(five)
zero = int(zero)
neg_8 = int(neg_8)
T = bool(T)
F = bool(F)

list3 = [five,zero,neg_8,T,F]

for x in list3:
  print(str(x)+str(type(x)))
print('\n')

#FOr some reason the last one converted to True
```

```
"""
Shopping List Calculator I
"""

# Create five variables,
# set them to strings that represent 5 common shopping list items


item_name_1 = "eggs"
item_name_2 = "milk"
item_name_3 = "butter"
item_name_4 = "bread"
item_name_5 = "cheese"

# Create five more variables,
# set them to floats that represent the prices of each of the items above

item_price_1 = 3.99
item_price_2 = 3.29
item_price_3 = 4.50
item_price_4 = 2.99
item_price_5 = 14.79

# Create five more variables,
# set them to ints that represent the quantity of each of the items above
item_quant_1 = 3
item_quant_2 = 1
item_quant_3 = 1
item_quant_4 = 2
item_quant_5 = 1

# Print to the console the name and price of each item defined above as follows:
# 1 Coco Puffs = $8.95.
# where:
# 1 would be item_quant_1
# Coco Puffs would be item_name_1
# 8.95 would be item_name_2

list2 = [0,1,2,3,4]

names = [item_name_1,item_name_2,item_name_3,item_name_4,item_name_5]
quantities = [item_quant_1,item_quant_2,item_quant_3,item_quant_4,item_quant_5]
prices = [item_price_1,item_price_2,item_price_3,item_price_4,item_price_5]

for x in list2:
  print(str(quantities[x])+" "+names[x]+' = '+str(prices[x]*quantities[x]))

```
