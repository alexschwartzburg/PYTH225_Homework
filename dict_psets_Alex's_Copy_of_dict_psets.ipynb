```
"""
Math with Girl Scout Cookies
"""

# Print out the number of boxes of girl scout cookies that each girl in the troop sold in the below format:
	# Wendy: _____
	# Connie: _____
	# Francesca: _____

Wendy = {'tagalongs': 5, 'thin mints': 12, 'samoas': 8}
Connie = {'tagalongs': 10, 'thin mints': 4, 'samoas': 12}
Francesca = {'tagalongs': 18, 'thin mints': 14, 'samoas': 10}

breakdown = list(Wendy)
print(breakdown)

#campers = [Wendy, Connie, Francesca]
Wendy_sales, Connie_sales, Francesca_sales = [] , [] , []
#camper_sales = [Wendy_sales, Connie_sales, Francesca_sales]

for x in breakdown: 
	y = Wendy[x]
	Wendy_sales.append(y)
print(Wendy_sales)

for x in breakdown: 
	y = Connie[x]
	Connie_sales.append(y)
print(Connie_sales)

for x in breakdown: 
	y = Francesca[x]
	Francesca_sales.append(y)
print(Francesca_sales)


salesW = sum(Wendy_sales)
salesC = sum(Connie_sales)
salesF = sum(Francesca_sales)
print('sales\n',
      'Wendy: ',salesW,'\n',\
      'Connie: ',salesC,'\n',\
			'Francesca: ',salesF)





# For each type of girl scout cookie, print out the total number of boxes sold in the below format:
	# tagalongs: _____
	# thin mints: _____
	# samoas: _____

total_tagalongs = Wendy['tagalongs']+Connie['tagalongs']+Francesca['tagalongs']
total_thinmints = Wendy['thin mints']+Connie['thin mints']+Francesca['thin mints']
total_samoas = Wendy['samoas']+Connie['samoas']+Francesca['samoas']

print('sales by cookie', '\n'\
		'tagalongs: ', total_tagalongs, '\n'\
		'thin mints:' , total_thinmints,'\n'\
	'samoas: ', total_samoas,'\n')

# For each type of girl scout cookie, print out the average number of boxes sold in the below format:
	# tagalongs: _____
	# thin mints: _____
	# samoas: _____

avg_tagalongs = (Wendy['tagalongs']+Connie['tagalongs']+Francesca['tagalongs'])/3
avg_thinmints = (Wendy['thin mints']+Connie['thin mints']+Francesca['thin mints'])/3
avg_samoas = (Wendy['samoas']+Connie['samoas']+Francesca['samoas'])/3

print('average sales by cookie', '\n'\
		'tagalongs: ', avg_tagalongs, '\n'\
		'thin mints:' , avg_thinmints,'\n'\
	'samoas: ', avg_samoas,'\n')

# Print out total the number of boxes of cookies the girls sold collectively as follows:
	# "This year we sold ______ boxes!"

boxes_sold = sum([salesW,salesC,salesF])
print(f"This year we sold {boxes_sold} boxes!")
```

```
"""
Inverting Keys & Values
"""

# Invert dict1 - make the current keys into values and the current values into keys.


dict1 = { "k1" : "v1", "k2" : "v2", "k3" : "v3" }
keys = []
values = []


for x in list(dict1):
  keys.append(x)

for x in keys:
  values.append(dict1[x])

dict2 = dict(zip(values, keys))
print(dict2)
```

```
"""
Grades
"""
# Here's a dict containing students' grades from the semester's assignments.

students = {
  'Ashton': [86, 45, 98, 100],
  'Sierra': [100, 100, 100, 100],
  'Zach': [38, 49, 90, 87],
  'Manuel': [98, 92, 86, 100],
  'Felicia': [94, 87, 89, 95],
  'Ankur': [75, 77, 77, 85],
  'Ananya': [98, 94, 87, 92],
  'Nick': [79, 84, 89, 90],
  'Olivia': [83, 91, 69, 85],
  'Molly': [83, 74, 72, 90]
}

# p 1.1) Create a dict called num_grades to store each student's average numerical grade for the semester.

student_names = list(students) # defines the list of students           
student_grades_avg = []       # creates the empty list

for x in student_names:    
  student_grades_avg.append(sum(students[x])/len(students[x]))
  #computes the avg grade of each student & then appends it to the empty list

num_grades = dict(zip(student_names, student_grades_avg)) # zips the two lists and creates a dictionary out of it



# p 1.2) Create a dict called letter_grades to store each student's final letter grade for the semester. 
# Use A, B, C, D, and F as grades per the standard grading scale.

cutoffs = [93, 90, 87, 83, 80, 77, 73, 70, 67, 63, 60, 59]
letters = ['A+','A','A-','B+','B','B-','C+','C','C-','D+','D','D-','F']
grades_numbers_to_letters = dict(zip(cutoffs, letters))

letter_grade = []
letter_grades = {}

# operations check
print('num_grades = ',num_grades)
print('letter_grade = ',letter_grade)







# #USED THIS TO COMPILE CONDITIONALS
# for x in cutoffs:
#   print(f'elif num_grades[x] >= {x}:\n\ty = grades_numbers_to_letters[{x}]\n\tprint(y)\n\tletter_grade.append(y))')

for x in student_names:
  if num_grades[x] >= 93:  # manually change 'elif' to 'if' here.
    y = grades_numbers_to_letters[93]
    letter_grade.append(y)
  elif num_grades[x] >= 90:
    y = grades_numbers_to_letters[90]
    letter_grade.append(y)
  elif num_grades[x] >= 87:
    y = grades_numbers_to_letters[87]
    letter_grade.append(y)
  elif num_grades[x] >= 83:
    y = grades_numbers_to_letters[83]
    letter_grade.append(y)
  elif num_grades[x] >= 80:
    y = grades_numbers_to_letters[80]
    letter_grade.append(y)
  elif num_grades[x] >= 77:
    y = grades_numbers_to_letters[77]
    letter_grade.append(y)
  elif num_grades[x] >= 73:
    y = grades_numbers_to_letters[73]
    letter_grade.append(y)
  elif num_grades[x] >= 70:
    y = grades_numbers_to_letters[70]
    letter_grade.append(y)
  elif num_grades[x] >= 67:
    y = grades_numbers_to_letters[67]
    letter_grade.append(y)
  elif num_grades[x] >= 63:
    y = grades_numbers_to_letters[63]
    letter_grade.append(y)
  elif num_grades[x] >= 60:
    y = grades_numbers_to_letters[60]
    letter_grade.append(y)
  elif num_grades[x] >= 59:
    y = grades_numbers_to_letters[59]
    letter_grade.append(y)

# operations check
print('letter grade = ',letter_grade)
print('letter_grades = ',letter_grades)


letter_grades.update(dict(zip(student_names, letter_grade)))
print('letter_grades = ',letter_grades)




# p 1.3) Create a list of honor roll students (i.e. students who got A's).

honor_roll = []

for x in student_names:
  if (letter_grades[x] == 'A+') or \
     (letter_grades[x] == 'A') or \
     (letter_grades[x] == 'A-'):
     honor_roll.append(x)
  else:
    pass

print(honor_roll)
```

```
"""
Price Inventory
"""

# Imagine you're the owner of an Italian restaurant. 
# You're evaluating your pricing to make sure it's up to date with the 
#   latest costs of making each menu item (see the menu prices and costs dicts below). 
# Evaluate whether each item's price returns a profit margin of at least $1.00. 
# If it doesn't, update the menu_prices dict with a new price that gives the item a $1.00 profit margin. 
# Then print out the items which you've repriced in a new dict called price_increases.

menu_prices = {
	'fettuccine bolognese': 14.99,
	'gnocchi gorgonzola': 10.99,
	'chicken parmigiano': 13.99,
	'butternut squash risotto': 9.99,
	'spaghetti carbonara': 12.99,
	'pizza margherita': 12.99,
	'chicken marsala': 13.99
}

menu_costs = {
	'fettuccine bolognese': 12.50,
	'gnocchi gorgonzola': 11.25,
	'chicken parmigiano': 11.75,
	'butternut squash risotto': 6.33,
	'spaghetti carbonara': 12.90,
	'pizza margherita': 14.00,
	'chicken marsala': 11.75
}

price_increases = {}

for x in list(menu_prices):
	if menu_prices[x] - (menu_costs[x]) < (1.00):
		price_increases.update({x:(1.00-((menu_prices[x])-(menu_costs[x])))}) 

print(price_increases)
#done here…

#…unless you want to round the price increases to the nearest penny, in which cas you need this block
price_increase = []
for x in price_increases:
	y = round(price_increases[x],2)
	price_increase.append(y)
price_increases.update(dict(zip(price_increases,price_increase)))

print(price_increases)
```

```
"""
Word Frequency
"""

# Print out the number of words in this movie quote. 
# Find and print out the most common word in the quote and how many times it was used.
### Hint: You do not need a loop for this. Look up the Counter docs in python3.

from collections import Counter

princess_bride = [
'Hello', 'my', 'name', 'is', 'Inigo', 'Montoya',
'You', 'killed', 'my', 'father',
'Prepare', 'to', 'die'
]

word_freq = Counter(princess_bride) 

#for x in word_freq:
#  print(x,':',word_freq[x])

words = list(word_freq)
freq = []
for x in words:
  freq.append(word_freq[x])

words_by_freq = dict(zip(freq,words))

fave_word = words_by_freq[max(freq)]
print(fave_word)

# p.s. You might use this to help analyze the most common topic in comments or reviews from your users to help understand the best places to improve you product.
```

```
"""
Summing Dict Values
"""

# Two Kindergarten teachers poll their classes for what fruit they want to eat for snacktime tomorrow. 
# Only one of them is going shopping, so she needs to know how many of each fruit she needs to buy in total. 
# Tally these up and assign them to the "shopping_list" dict.

poll1 = {'apples': 8, 'bananas': 12}
poll2 = {'apples': 6, 'bananas': 6, 'clementines': 8}

polls = list([poll1, poll2])

items = []
for x in list(poll1):    #dump the items from 1st poll
  items.append(x)
for x in list(poll2):    #dump the items from 2nd poll
  items.append(x)
count = []               #create a new list
for x in set(items):     #cycle through that SET of items
  try:                   #handle the lack of perfect overlap between poll1 and poll2
    y = poll1[x]
  except Exception:
    y = 0
  try:
    z = poll2[x]
  except Exception:
    z = 0
  count.append(y+z)      #append the sum of each item to the new list

shopping_list = dict(zip(list(set(items)),count)) #zip the two created lists as a dictionary


print(shopping_list)

```
